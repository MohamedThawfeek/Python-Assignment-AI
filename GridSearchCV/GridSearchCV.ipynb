```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
dataset=pd.read_csv("insurance_pre.csv")
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)

Input=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
Output=dataset[['charges']]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Input, Output, test_size=1/3, random_state=0)


```

```python
from sklearn.svm import SVR

params = {"kernel": ["linear", "poly", "rbf", "sigmoid"], "C": [10, 100, 1000, 2000, 3000], "gamma": ["scale", "auto"]}
model = GridSearchCV(SVR(), params, refit=True, verbose=3, n_jobs=1)
model.fit(x_train, y_train)

y_pred=model.predict(x_test)
from sklearn.metrics import r2_score
score =r2_score(y_test, y_pred)

print("Get the best model {}:".format(model.best_params_), score)
```

```python
from sklearn.tree import DecisionTreeRegressor

params2 = {"criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"], "splitter": ["best", "random"], "max_features": ["sqrt", "log2"]}
model2 = GridSearchCV(DecisionTreeRegressor(), params2, refit=True)
model2.fit(x_train, y_train)

y_pred2=model2.predict(x_test)

from sklearn.metrics import r2_score
score2 =r2_score(y_test, y_pred2)

print("Get the best model {}:".format(model2.best_params_), score2)
```

```python
from sklearn.ensemble import RandomForestRegressor

params3 = {"criterion": ["squared_error", "absolute_error", "poisson", "friedman_mse"], "n_estimators": [10, 100], "max_features": ["sqrt", "log2",]}
model3 = GridSearchCV(RandomForestRegressor(), params3, refit=True)
model3.fit(x_train, y_train)

y_pred3=model3.predict(x_test)
from sklearn.metrics import r2_score
score3 =r2_score(y_test, y_pred3)

print("Get the best model {}:".format(model3.best_params_), score3)
```

```python
from sklearn.ensemble import AdaBoostRegressor

params4 = {"n_estimators": [10, 100]}
model4 = GridSearchCV(AdaBoostRegressor(), params4, refit=True)
model4.fit(x_train, y_train)

y_pred4=model4.predict(x_test)
from sklearn.metrics import r2_score
score4 =r2_score(y_test, y_pred4)

print("Get the best model {}:".format(model4.best_params_), score4)
```
