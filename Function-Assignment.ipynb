```python
# Create a function that checks whether the given number is Odd or Even
def OddEven():
    number= int(input("Enter a number:"))
    if ((number%2) ==1):
        print(number, "is an Odd Number")
    else:
        print(number, "is an Event Number")

OddEven()
```
