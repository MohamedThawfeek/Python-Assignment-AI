```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.cluster import KMeans

dataset=pd.read_csv("Mall_Customers.csv")
Input=dataset.iloc[:, 3:5]
array_list=[]

for i in range(1,11):
    model=KMeans(n_clusters=i, init='k-means++', random_state=42)
    model.fit(Input)
    array_list.append(model.inertia_)

plt.plot(range(1,11), array_list)
plt.title("The Elbow method")
plt.xlabel("Clustring")
plt.ylabel("WCSS")
plt.show()

model=KMeans(n_clusters=5, init='k-means++', random_state=42)
y_predict= model.fit_predict(Input)

supervice_table=pd.DataFrame(dataset)
supervice_table["Cluster_group"]=y_predict 

cluster_chart= sbn.lmplot(data=supervice_table, x=supervice_table.columns[3], y=supervice_table.columns[4],
                          hue=supervice_table.columns[5], fit_reg=False, legend=True, legend_out=True)

    
```
