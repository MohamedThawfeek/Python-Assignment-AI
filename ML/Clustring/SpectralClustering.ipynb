```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score  # Using silhouette score instead of inertia

dataset=pd.read_csv("Mall_Customers.csv")
Input=dataset.iloc[:, 3:5]
array_list=[]

model=SpectralClustering(n_clusters=2, assign_labels='discretize', random_state=0)
y_predict= model.fit_predict(Input)

y_predict
supervice_table=pd.DataFrame(dataset)
supervice_table["Cluster_group"]=y_predict 

score = silhouette_score(Input, y_predict)
print(score)

cluster_chart= sbn.lmplot(data=supervice_table, x=supervice_table.columns[3], y=supervice_table.columns[4],
                          hue=supervice_table.columns[5], fit_reg=False, legend=True, legend_out=True)

    
```
