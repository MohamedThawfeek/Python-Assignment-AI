```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score

dataset= pd.read_csv("CKD.csv")
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset['age'] = dataset['age'].round(0).astype(int)
dataset['bp'] = dataset['bp'].round(2)

dataset['bgr'] = dataset['bgr'].round(2)
dataset['bu'] = dataset['bu'].round(2)
dataset['sc'] = dataset['sc'].round(2)
dataset['sod'] = dataset['sod'].round(2)
dataset['pot'] = dataset['pot'].round(2)
dataset['hrmo'] = dataset['hrmo'].round(2)
dataset['pcv'] = dataset['pcv'].round(2)
dataset['wc'] = dataset['wc'].round(2)
dataset['rc'] = dataset['rc'].round(2)


Input=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
Output=dataset[['classification_yes']]




count=dataset[['classification_yes']].value_counts()
count

x_train, x_test, y_train, y_test = train_test_split(Input,Output, test_size=1/3, random_state=0)


```

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression

params={'penalty': ['l1', 'l2', 'elasticnet', None], 'multi_class': ['auto', 'ovr', 'multinomial']}
model= GridSearchCV(LogisticRegression(), params, refit=True, verbose=3, n_jobs=1)
model.fit(x_train, y_train)
y_pred= model.predict(x_test)

confusion_score = confusion_matrix(y_test, y_pred)
report_score=classification_report(y_test, y_pred)
roc_score=roc_auc_score(y_test,model.predict_proba(x_test)[:, 1])

print("Get the best model 01", format(model.best_params_))
print(f'Roc Model Score01: {roc_score:.2f}')
print(report_score)
```

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

params={'criterion': ['gini', 'entropy', 'log_loss'], 'splitter': ['best', 'random'], 'random_state': [0]}
model= GridSearchCV(DecisionTreeClassifier(), params, refit=True, verbose=3, n_jobs=1, scoring="f1_weighted")
model.fit(x_train, y_train)
y_pred= model.predict(x_test)

confusion_score = confusion_matrix(y_test, y_pred)
report_score=classification_report(y_test, y_pred)
roc_score=roc_auc_score(y_test,model.predict_proba(x_test)[:, 1])

print("Get the best model 01", format(model.best_params_))
print(f'Roc Model Score01: {roc_score:.2f}')
print(report_score)
print(confusion_score)
```

```python
#RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
import pickle

params={'n_estimators': [10,50,100], 'criterion': ['gini', 'entropy', 'log_loss'], 'max_features': ['sqrt', 'log2', None], 'random_state': [0]}
model= GridSearchCV(RandomForestClassifier(), params, refit=True, verbose=3, n_jobs=1, scoring="f1_weighted")
model.fit(x_train, y_train)
re=model.cv_results_
y_pred= model.predict(x_test)




confusion_score = confusion_matrix(y_test, y_pred)
report_score=classification_report(y_test, y_pred)
roc_score=roc_auc_score(y_test,model.predict_proba(x_test)[:, 1])

pickle.dump(model, open("best_model_classification.sav", "wb"))



print("Get the best model 01", format(model.best_params_))
print(f'Roc Model Score01: {roc_score:.2f}')
print(report_score)
print(confusion_score)
```
