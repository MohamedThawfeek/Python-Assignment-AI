```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
import pickle

dataset= pd.read_csv("Social_Network_Ads.csv")
dataset=pd.get_dummies(dataset,dtype=int, drop_first=True)
dataset=dataset.drop('User ID', axis=1)


Input=dataset[['Age', 'EstimatedSalary','Gender_Male']]
Output=dataset[['Purchased']]

dataset[['Purchased']].value_counts()

x_train,x_test,y_train,y_test = train_test_split(Input,Output, test_size=0.3, random_state=0)

params={'n_estimators': [10,50,100] ,'criterion': ['gini', 'entropy', 'log_loss'], 'max_features': ['sqrt', 'log2', None], 'random_state': [0]}
model = GridSearchCV(RandomForestClassifier(), params, refit=True, verbose=3, n_jobs=1)
model.fit(x_train, y_train)
y_pred= model.predict(x_test)

score = confusion_matrix(y_test, y_pred)
score2=classification_report(y_test, y_pred)

pickle.dump(model,open("best_model_RFC.sav", "wb"))

print("Get the best model", format(model.best_params_))
print(score2)






```
