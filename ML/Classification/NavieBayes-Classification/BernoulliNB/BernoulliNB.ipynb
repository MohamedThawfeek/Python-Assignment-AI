```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
import pickle

dataset= pd.read_csv("Social_Network_Ads.csv")
dataset=pd.get_dummies(dataset,dtype=int, drop_first=True)
dataset=dataset.drop('User ID', axis=1)


Input=dataset[['Age', 'EstimatedSalary','Gender_Male']]
Output=dataset[['Purchased']]

dataset[['Purchased']].value_counts()

x_train,x_test,y_train,y_test = train_test_split(Input,Output, test_size=0.3, random_state=0)

params={}
model = GridSearchCV(BernoulliNB(), params, refit=True, verbose=3, n_jobs=1, scoring='f1_weighted')
model.fit(x_train, y_train)
y_pred= model.predict(x_test)

score = confusion_matrix(y_test, y_pred)
score2=classification_report(y_test, y_pred)

score3=roc_auc_score(y_test,model.predict_proba(x_test)[:, 1])


pickle.dump(model,open("best_model_BNB.sav", "wb"))

print("Get the best model", format(model.best_params_))
print(f'Model score:, {score3:.2f}')
print(score2)






```
