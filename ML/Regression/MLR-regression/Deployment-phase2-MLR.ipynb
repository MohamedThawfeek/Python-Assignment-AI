```python
import pickle

load_model=pickle.load(open("final_model.sav", "rb"))
r_d_spend=float(input("Spend money for R&D:"))
administration=float(input("Spend money for administration:"))
marketing=float(input("Spend money for marketing:"))
state=int(input("Spend money for state:"))
state1=int(input("Spend money for state1:"))
result = load_model.predict([[r_d_spend, administration, marketing, state, state1]])

print(f"Predicted Profit: ₹{int(result)}")
```
