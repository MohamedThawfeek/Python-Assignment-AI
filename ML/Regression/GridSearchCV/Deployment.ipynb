```python
import pickle
load_model = pickle.load(open("best_model.sav", "rb"))

def calculate_bmi(weight_kg, height_feet):
    # Convert height from feet to meters
    height_meters = height_feet * 0.3048
    # Calculate BMI
    bmi = weight_kg / (height_meters ** 2)
    return float(f"{bmi:.2f}")

def gendervalue(gender):
    if (gender.lower() == "male"):
        response = 1
    else:
        response = 0 

    return response


def smokervalue(smoker):
    if (smoker.lower() == "yes"):
        smokerResponse = 1
    else:
        smokerResponse = 0 

    return smokerResponse


age = int(input("Enter your age: "))
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))  # Input in meters
children= int(input("Enter how many children do you have: "))
gender = input("Enter your gender: ")
smoker = input("Do you smoke: Give the value Yes/No: ")

Gender=gendervalue(gender)
Smoker=smokervalue(smoker)
BMIValue = calculate_bmi(weight, height)

result = load_model.predict([[age, BMIValue, children, Gender, Smoker]])

print(f"Predicted Profit: ₹{int(result)}")
```
