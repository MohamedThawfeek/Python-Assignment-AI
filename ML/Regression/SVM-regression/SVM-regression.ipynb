```python
import pandas as pd
dataset= pd.read_csv("50_Startups.csv")
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
Input=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
Output=dataset[['Profit']]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Input, Output, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)

from sklearn.metrics import r2_score
score=r2_score(y_test, y_pred)
score
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVR
#params (linear, rbf, poly, sigmoid, (precomputed use case is 35x35))
model1=SVR(kernel="sigmoid", C=10)
model1.fit(x_train, y_train)
y_pred1=model1.predict(x_test)

from sklearn.metrics import r2_score
score1=r2_score(y_test, y_pred1)
score1
```

```python
from sklearn.tree import DecisionTreeRegressor
#Params (“squared_error”, “friedman_mse”, “absolute_error”, “poisson”)
#Spliter params (random, best)
# max_features {“sqrt”, “log2”},
model2=DecisionTreeRegressor(criterion="absolute_error", splitter="random", max_features="log2" )
model2.fit(x_train, y_train)

# import matplotlib.pyplot as plt
# from sklearn import tree
# tree.plot_tree(model2)
# plt.show()

  
y_pred2=model2.predict(x_test)

from sklearn.metrics import r2_score
score2=r2_score(y_test, y_pred2)
score2
```

```python
from sklearn.ensemble import RandomForestRegressor
#Params criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”},
# max_features {“sqrt”, “log2”},
#n_estimators 10 to 100
model3=RandomForestRegressor(criterion="squared_error", n_estimators=50, random_state=0)
model3.fit(x_train, y_train)
  
y_pred3=model3.predict(x_test)

from sklearn.metrics import r2_score
score3=r2_score(y_test, y_pred3)
score3
```
