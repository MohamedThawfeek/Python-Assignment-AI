```python
import pandas as pd

dataset=pd.read_csv("insurance_pre.csv")
dataset=pd.get_dummies(dataset,dtype=int, drop_first=True)

Input=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
Output=dataset[['charges']]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Input,Output, test_size=0.3, random_state=0)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)



y_pred=model.predict(x_test)

from sklearn.metrics import r2_score
score=r2_score(y_test, y_pred)

score
```

```python
#SVM

#params (linear, rbf, poly, sigmoid, (precomputed use case is 35x35))

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.svm import SVR
model2=SVR(kernel="sigmoid", C=10)
model2.fit(x_train, y_train)



y_pred2=model2.predict(x_test)

from sklearn.metrics import r2_score
score2=r2_score(y_test, y_pred2)

score2
```

```python
#DT

#params criterion{“squared_error”, “friedman_mse”, “absolute_error”, “poisson”},
# splitter{“best”, “random”}, default=”best”
# max_featuresint {“sqrt”, “log2”}, default=None

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.tree import DecisionTreeRegressor
model3=DecisionTreeRegressor(criterion='poisson', splitter='random', max_features="log2")
model3.fit(x_train, y_train)



y_pred3=model3.predict(x_test)

from sklearn.metrics import r2_score
score3=r2_score(y_test, y_pred3)

score3
```

```python
#RF

#params criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”}
# max_features {“sqrt”, “log2”}, default=None

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.ensemble import RandomForestRegressor
model4=RandomForestRegressor(criterion="absolute_error", n_estimators=50, random_state=0, max_features="sqrt")
model4.fit(x_train, y_train)



y_pred4=model4.predict(x_test)

from sklearn.metrics import r2_score
score4=r2_score(y_test, y_pred4)

score4
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

params = {"kernel": ["linear", "poly", "rbf", "sigmoid"], "C": [10, 100, 1000, 2000, 3000], "gamma": ["scale", "auto"]}
grid_model = GridSearchCV(SVR(), params, refit=True,verbose=3, n_jobs=1)

grid_model.fit(x_train, y_train)

grid_predict = grid_model.predict(x_test)

from sklearn.metrics import r2_score
score = r2_score(y_test,grid_predict)

print("Get the best model", format(grid_model.best_params_), score)

```
