```python
import pandas as pd
dataset=pd.read_csv("50_Startups.csv")
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)

Input=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
Output=dataset[['Profit']]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Input, Output, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)

y_pred=model.predict(x_test)

from sklearn.metrics import r2_score
score=r2_score(y_test, y_pred)

score
```

```python
import pickle
pickle.dump(model, open("final_model.sav", "wb"))
```
